### Import Data

```python
import pandas as pd
from _classification._utilities_classification import *
```

Full Features

```python
df_train = pd.read_csv("Datasets Training.csv")
df_test = pd.read_csv("Datasets Testing.csv")

X_train, y_train  = preprocess(df_train)
X_test, y_test  = preprocess(df_test)

print(X_train.shape)
print(X_test.shape)
```

Selected Features

```python
df_train_s = pd.read_csv("Datasets Training (selected).csv")
df_test_s = pd.read_csv("Datasets Testing (selected).csv")

X_train_s, y_train_s  = preprocess(df_train_s)
X_test_s, y_test_s  = preprocess(df_test_s)

print(X_train_s.shape)
print(X_test_s.shape)
```

### Classification

```python
###################################################################################
# Classifier
from sklearn.neural_network import MLPClassifier
# from sklearn.svm import SVC #SVM Classifier
# from sklearn.naive_bayes import GaussianNB # Gaussian Naive Bayer Classifier
###################################################################################

clf = MLPClassifier(hidden_layer_sizes=53, max_iter=1000, random_state=1)
```

```python
train_features, test_features, train_labels, test_labels = X_train, X_test, y_train, y_test

# Train MLP
clf.fit(train_features, train_labels)
y_pred = clf.predict(test_features)
accuracy = accuracy_score(test_labels, y_pred)
accuracy
```
