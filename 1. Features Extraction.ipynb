Import Images

```python
from __utils__.import_file import import_images
```

```python
# Import data train
path_train_positif = r"Images\Train\Positif"
path_train_negatif = r"Images\Train\Negatif"

train_images_positif, labels_train_images_positif = import_images(path_train_positif)
train_images_negatif, labels_train_images_negatif = import_images(path_train_negatif)

data_train = []
labels_train = []

data_train.extend(train_images_positif)
data_train.extend(train_images_negatif)
labels_train.extend(labels_train_images_positif)
labels_train.extend(labels_train_images_negatif)

print("sum of data training: " + str(len(data_train)))
print("sum of labels training: " + str(len(labels_train)))
```

```python
# Import data test
path_test_positif = r"Images\Test\Positif"
path_test_negatif = r"Images\Test\Negatif"

test_images_positif, labels_test_images_positif = import_images(path_test_positif)
test_images_negatif, labels_test_images_negatif = import_images(path_test_negatif)

data_test = []
labels_test = []

data_test.extend(test_images_positif)
data_test.extend(test_images_negatif)
labels_test.extend(labels_test_images_positif)
labels_test.extend(labels_test_images_negatif)

print("sum of data testing: " + str(len(data_test)))
print("sum of labels testing: " + str(len(labels_test)))
```

Features Extraction

```python
from _features.features_extraction import *

train = extract_texture(data_train, mask=None)
test = extract_texture(data_test, mask=None)
```

```python
train.GLCM_mean
```

Convert to Dataframe

```python
from __utils__.convert2dataframe import *
```

```python
df_train = convert2dataframe(train, labels_train)

df_result_training = pd.concat([df_train.fos, 
                                df_train.GLCM_mean, 
                                df_train.GLCM_four_dir,
                                df_train.GLDS, 
                                df_train.GLRLM, 
                                df_train.Class], axis=1)
df_result_training
```

```python
df_test = convert2dataframe(test, labels_test)

df_result_testing = pd.concat([df_test.fos, 
                                df_test.GLCM_mean, 
                                df_test.GLCM_four_dir,
                                df_test.GLDS, 
                                df_test.GLRLM, 
                                df_test.Class], axis=1)
df_result_testing
```

Export Dataframe

```python
df_result_training.to_csv("Datasets Training.csv", index=False)
df_result_testing.to_csv("Datasets Testing.csv", index=False)
```
